Nota-se que para os campos de classe trabalhadora e ocupação, existe a alternativa "?" <br>
Para que a visualização dos dados fique mais simples, os dados de maior relevância podem ser classificados de forma binária e distribuidos em um número maior de colunas, da seguinte forma:

```python
# Visualização dos dados de ocupação de forma binária
occupation = pd.get_dummies(df.occupation).add_prefix('occupation_').astype('int')
occupation
```

```python
# Concatenação dos dados binários com o dataframe original
df = df.drop('occupation', axis=1) # Remoção da coluna occupation
df = pd.concat([df, occupation], axis=1) # Concatenação dos dados binários com o dataframe original
df 
```

Fazendo o mesmo para os demais parâmetros, tem-se:

```python
# Visualização dos dados de workclass de forma binária
workclass = pd.get_dummies(df.workclass).add_prefix('workclass_').astype('int')
# Concatenação dos dados binários com o dataframe original
df = df.drop('workclass', axis=1) # Remoção da coluna classe trabalhadora
df = pd.concat([df, workclass], axis=1) # Concatenação dos dados binários com o dataframe original

# Visualização dos dados de estado civil de forma binária
maritalstatus = pd.get_dummies(df['marital-status']).add_prefix('marital-status_').astype('int')
# Concatenação dos dados binários com o dataframe original
df = df.drop('marital-status', axis=1) # Remoção da coluna estado civil
df = pd.concat([df, maritalstatus], axis=1) # Concatenação dos dados binários com o dataframe original

# Visualização dos dados de relacionamento de forma binária
relationship = pd.get_dummies(df.relationship).add_prefix('relationship_').astype('int')
# Concatenação dos dados binários com o dataframe original
df = df.drop('relationship', axis=1) # Remoção da coluna relationship
df = pd.concat([df, relationship], axis=1) # Concatenação dos dados binários com o dataframe original

# Visualização dos dados de país de origem de forma binária
nativecountry = pd.get_dummies(df['native-country']).add_prefix('native-country_').astype('int')
# Concatenação dos dados binários com o dataframe original
df = df.drop('native-country', axis=1) # Remoção da coluna de país de origem
df = pd.concat([df, nativecountry], axis=1) # Concatenação dos dados binários com o dataframe original

# Visualização dos dados de etnia de forma binária
race = pd.get_dummies(df['race']).add_prefix('race_').astype('int')
# Concatenação dos dados binários com o dataframe original
df = df.drop('race', axis=1) # Remoção da coluna de etnia
df = pd.concat([df, race], axis=1) # Concatenação dos dados binários com o dataframe original

# Remoção da coluna 'education', já que o nível de educação é representado pela coluna 'educational-num'
df = df.drop('education', axis=1) # Remoção da coluna de education

# Visualização do dataframe após a transformação dos dados
df 
```

Nota-se que os dados de renda ('income') e gênero ('gender') já são dados binários no dataframe, uma vez que renda assume os valores <=50K ou >50K e gênero é dado como maculino ou feminino. <br>
Nesse caso, podemos fazer uma categorização binária assumindo que: <br>
##### Renda:
0 para renda <=50K; <br>
1 para renda >50K; <br>
##### Gênero:
1 para masculino; <br>
0 para feminino; <br>

```python
df['income'] = df['income'].apply(lambda x: 1 if x == ">50K" else 0)
df['gender'] = df['gender'].apply(lambda x: 1 if x == "Male" else 0)

df
```

Agora, os dados podem ser visualizados num mapa de calor para visualizar a correlação entre as variáveis para a definição da renda:

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))
sns.heatmap(df.corr(), annot=False, cmap = 'coolwarm')
```

Como o parâmetro de interesse é a renda, os dados podem ser filtrados para que seja exibido apenas os que estão mais correlacionados com a renda:

```python
correlations = df.corr()['income'].abs()
sorted_correlations = correlations.sort_values()
num_cols_to_drop = int (0.8 * len(df.columns))
cols_to_drop = sorted_correlations.iloc[:num_cols_to_drop].index
df_dropped = df.drop(cols_to_drop, axis=1)

df_dropped
```

Como o dataset segue um padrão binário, pode-se utilizar um algoritmo de machine learning de arvore binária, como Random Forest:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Separação dos dados de treino e teste
# O valor 0.2 indica que 20% dos dados serão utilizados para teste
# O valor 42 é um número arbitrário utilizado para garantir a reprodutibilidade dos resultados

df = df.drop('fnlwgt', axis=1) # Remoção da coluna fnlwgt, já que ela não é relevante para o modelo
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

train_df
```

```python
test_df
```

O código abaixo separa os dados de treino e teste em duas variáveis diferentes. 
<br>
`train_X` é um dataframe que contém todas as colunas do dataframe `train_df`, exceto a coluna `income`. Ou seja, `train_X` contém todas as variáveis independentes que serão utilizadas para prever a variável dependente `income`.
<br>
`train_Y` é uma série que contém apenas a coluna `income` do dataframe `train_df`. Ou seja, `train_Y` contém a variável dependente que será utilizada para treinar o modelo.
<br>
`test_X` é um dataframe que contém todas as colunas do dataframe `test_df`, exceto a coluna `income`. Ou seja, `test_X` contém todas as variáveis independentes que serão utilizadas para prever a variável dependente `income` no conjunto de teste.
<br>
`test_Y` é uma série que contém apenas a coluna `income` do dataframe `test_df`. Ou seja, `test_Y` contém a variável dependente que será utilizada para avaliar o desempenho do modelo no conjunto de teste.

```python
train_X = train_df.drop('income', axis=1)
train_Y = train_df['income']

test_X = test_df.drop('income', axis=1)
test_Y = test_df['income']
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state=42)

forest.fit(train_X, train_Y)
```

```python
# Verificação da precisão do método
forest.score(test_X, test_Y)
```

```python
# Verificação da importância de cada feature (variável independente) no modelo de Random Forest treinado
importances = dict(zip(forest.feature_names_in_,forest.feature_importances_))
importances = {k: v for k, v in sorted(importances.items(), key=lambda x: x[1], reverse=True)}
importances
```

O código abaixo é uma implementação do algoritmo Grid Search para encontrar os melhores hiperparâmetros para o modelo de Random Forest. <br>
O Grid Search é uma técnica de busca exaustiva que testa todas as combinações possíveis de hiperparâmetros especificados no dicionário `param_grid`. <br>
Nesse caso, o dicionário `param_grid` contém as seguintes opções de hiperparâmetros para o modelo de Random Forest: <br>
- `n_estimators`: número de árvores na floresta; <br>
- `max_depth`: profundidade máxima da árvore; <br>
- `min_samples_split`: número mínimo de amostras necessárias para dividir um nó interno; <br>
- `max_features`: número máximo de features a serem consideradas em cada divisão. <br>
O  `GridSearchCV` recebe como parâmetros o modelo a ser utilizado (`RandomForestClassifier`), o dicionário de hiperparâmetros (`param_grid`) e o parâmetro `verbose=10`, que indica o nível de detalhamento das informações exibidas durante a execução do Grid Search. <br><br>

Após a execução do Grid Search, `grid_search` contém os melhores hiperparâmetros encontrados para o modelo de Random Forest.

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [50, 100, 250],
    'max_depth': [5, 10, 30, None],
    'min_samples_split': [2, 4],
    'max_features': ['sqrt', 'log2'],
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid=param_grid,verbose=10)
```

```python
# Treinamento do modelo
grid_search.fit(train_X, train_Y)
```

```python
grid_search.best_estimator_
```

```python
# Verificação da precisão do método com os melhores parâmetros
forest = grid_search.best_estimator_
forest.score(test_X, test_Y)
```

```python
# Verificação da importância de cada feature (variável independente) no modelo de Random Forest treinado com os melhores parâmetros
importances = dict(zip(forest.feature_names_in_,forest.feature_importances_))
importances = {k: v for k, v in sorted(importances.items(), key=lambda x: x[1], reverse=True)}
importances
```
